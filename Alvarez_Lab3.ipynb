$$NV = a*vect_1 + b*vect_2 $$
$$v_1 = \begin{bmatrix} 4\\9 \\\end{bmatrix} , v_2 = \begin{bmatrix} -5\\7 \\\end{bmatrix} $$

```python
vect1 = np.array([4,9])
vect2 = np.array([-5,7])
a = 1
b = 2 
NV = a*vect1 + b*vect2

print(NV)
```

```python
vectA = np.array([4,9])
vectB = np.array([-5,7])
R = np.arange(-10,10,1)
v1, v2 = np.meshgrid(R,R)
vectR = vectA + vectB
spanRx = v1*vectA[0] + v2*vectB[0]
spanRy = v1*vectA[1] + v2*vectB[1]
plt.scatter(spanRx,spanRy, s=5, alpha=0.75)

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.grid()
plt.show()
```

$$Z= \begin{Bmatrix} f_1 \cdot\begin{bmatrix} 3\\1 \\\end{bmatrix},
